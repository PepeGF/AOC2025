```python
with open('input.txt', 'r') as file:
    data = file.read()

rangos: list[tuple[int, int]] = []
partes = data.split(",")
for parte in partes:
    temp = parte.split("-")
    rangos.append((int(temp[0]), int(temp[1])))
```
